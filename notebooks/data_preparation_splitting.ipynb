```python
df_station=pd.read_excel('../data/data reference/codification clients.xlsx',header=1)
df_station
```

```python
df_fusionné = pd.merge(df_stations_service, df_station[['NUMCPT', 'STATION']], on='NUMCPT', suffixes=['_ventes', '_stations'])

# Afficher le dataframe fusionné
df_fusionné
```

```python
df_fusionné['STATION'].value_counts()
```

```python
import re

stations = df_fusionné['STATION'].unique()
def clean_filename(filename):
    return re.sub(r'[<>:"/\\|?*]', '_', filename)
# Créer un fichier CSV pour chaque station
for station in stations:
    # Filtrer les données pour la station actuelle
    df_chaque_station = df_fusionné[df_fusionné['STATION'] == station]
    
    # Créer un nom de fichier basé sur le nom de la station (remplacer les espaces par des underscores et nettoyer les caractères invalides)
    station_filename = clean_filename(station.replace(' ', '_').replace('.', '').replace('/', '_'))
    filename = f"{station_filename}.csv"
    
    # Sauvegarder les données dans un fichier CSV
    df_chaque_station.to_csv(f"../data/data each station_service/{filename}", index=False)

    print(f"Fichier {filename} créé avec {len(df_chaque_station)} lignes.")
```
